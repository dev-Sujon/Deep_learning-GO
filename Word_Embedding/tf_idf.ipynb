### Overview of TF-IDF

TF-IDF is a statistical measure used to evaluate the importance of a word in a document relative to a collection of documents (corpus). The intuition behind TF-IDF is that if a word appears frequently in a document but not in many documents within the corpus, it should be given a higher weight.

### Components of TF-IDF

1. **Term Frequency (TF)**: This measures the frequency of a term in a document. The more frequently a term appears in a document, the higher its TF.

   \[
   \text{TF}(t, d) = \frac{\text{Number of times term } t \text{ appears in document } d}{\text{Total number of terms in document } d}
   \]

   For example, if the term "machine" appears 3 times in a document containing 100 words, the term frequency would be \( \frac{3}{100} = 0.03 \).

2. **Inverse Document Frequency (IDF)**: This measures the importance of a term across the entire corpus. Terms that appear in many documents have a lower IDF score.

   \[
   \text{IDF}(t, D) = \log \left( \frac{N}{1 + \text{DF}(t)} \right)
   \]

   where \( N \) is the total number of documents and \( \text{DF}(t) \) is the number of documents containing the term \( t \). The "+1" is added to prevent division by zero for terms that do not appear in any document.

   For example, if the term "machine" appears in 10 out of 1000 documents, the inverse document frequency would be \( \log \left( \frac{1000}{1 + 10} \right) \approx 2.00 \).

3. **TF-IDF Score**: The TF-IDF score is the product of TF and IDF, providing a measure of how important a term is within a document in the context of the entire corpus.

   \[
   \text{TF-IDF}(t, d, D) = \text{TF}(t, d) \times \text{IDF}(t, D)
   \]

### Calculating TF-IDF: An Example

Consider a simple corpus with three documents:

1. **Documents**:
   ```
   D1: "the cat in the hat"
   D2: "the cat is on the mat"
   D3: "the mat is in the hat"
   ```

2. **Vocabulary**:
   ```
   ["the", "cat", "in", "hat", "is", "on", "mat"]
   ```

3. **Term Frequency (TF)**:
   - **D1**: [2, 1, 1, 1, 0, 0, 0]
   - **D2**: [2, 1, 0, 0, 1, 1, 1]
   - **D3**: [2, 0, 1, 1, 1, 0, 1]

4. **Document Frequency (DF)**:
   ```
   ["the" (3), "cat" (2), "in" (2), "hat" (2), "is" (2), "on" (1), "mat" (2)]
   ```

5. **Inverse Document Frequency (IDF)**:
   \[
   \text{IDF}(t) = \log \left( \frac{N}{1 + \text{DF}(t)} \right)
   \]
   \[
   \text{IDF}(\text{"the"}) = \log \left( \frac{3}{1 + 3} \right) \approx 0
   \]
   \[
   \text{IDF}(\text{"cat"}) = \log \left( \frac{3}{1 + 2} \right) \approx 0.405
   \]
   \[
   \text{IDF}(\text{"in"}) = \log \left( \frac{3}{1 + 2} \right) \approx 0.405
   \]
   \[
   \text{IDF}(\text{"hat"}) = \log \left( \frac{3}{1 + 2} \right) \approx 0.405
   \]
   \[
   \text{IDF}(\text{"is"}) = \log \left( \frac{3}{1 + 2} \right) \approx 0.405
   \]
   \[
   \text{IDF}(\text{"on"}) = \log \left( \frac{3}{1 + 1} \right) \approx 1.098
   \]
   \[
   \text{IDF}(\text{"mat"}) = \log \left( \frac{3}{1 + 2} \right) \approx 0.405
   \]

6. **TF-IDF**:
   - **D1**: [0, 0.405, 0.405, 0.405, 0, 0, 0]
   - **D2**: [0, 0.405, 0, 0, 0.405, 1.098, 0.405]
   - **D3**: [0, 0, 0.405, 0.405, 0.405, 0, 0.405]


### Applications of TF-IDF

1. **Information Retrieval**: TF-IDF is widely used in search engines to rank documents based on their relevance to a query.
2. **Text Mining**: It helps in extracting significant terms from documents for analysis.
3. **Document Clustering**: TF-IDF vectors can be used to cluster similar documents together.
4. **Content Recommendation**: TF-IDF is used in recommendation systems to suggest relevant content based on term importance.
5. **Spam Detection**: It helps in identifying spam emails by analyzing term frequencies and patterns.

### Advantages of TF-IDF

- **Simple and Effective**: Easy to implement and effective for many NLP tasks.
- **Noise Reduction**: Reduces the impact of common but less informative words.
- **Sparse Representation**: Efficiently represents documents in a high-dimensional space.

### Limitations of TF-IDF

- **Lack of Semantic Understanding**: TF-IDF does not capture semantic relationships between words.
- **Vocabulary Size**: The dimensionality of TF-IDF vectors is equal to the size of the vocabulary, which can be very large.
- **Context Ignorance**: Treats each word independently, ignoring word order and context.

### Comparison with Word Embeddings

While TF-IDF is a powerful tool for text representation, it differs significantly from word embeddings like Word2Vec, GloVe, and BERT:

- **Dimensionality**: TF-IDF vectors have the same dimension as the vocabulary size, while word embeddings typically have fixed, lower dimensions (e.g., 100, 300).
- **Contextual Information**: Word embeddings capture semantic and contextual information, whereas TF-IDF treats each word independently.
- **Sparsity**: TF-IDF results in sparse vectors, whereas word embeddings produce dense vectors.
- **Training Data**: TF-IDF is based on term frequencies within a specific corpus, whereas word embeddings are pre-trained on large datasets to capture general linguistic patterns.

### Conclusion

TF-IDF is a fundamental technique in NLP for representing the importance of terms in documents relative to a corpus. Its simplicity and effectiveness make it a popular choice for many text analysis tasks. However, for applications requiring deeper semantic understanding, word embeddings provide more sophisticated representations. Understanding both TF-IDF and word embeddings allows you to choose the right tool for your specific NLP tasks.


### Implementing TF-IDF in Python with Scikit-Learn

````python
from sklearn.feature_extraction.text import CountVectorizer
corpus = [
    "the cat in the hat",
    "the cat is on the mat",
    "the mat is in the hat"
]
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(corpus)

print(vectorizer.get_feature_names_out())
print(X.toarray())
````

````python
# prompt: visualize

import matplotlib.pyplot as plt

# Get feature names
feature_names = vectorizer.get_feature_names_out()

# Get TF-IDF values
tf_idf_values = X.toarray()

# Create a bar chart for each document
for i, doc in enumerate(corpus):
    plt.figure()
    plt.bar(feature_names, tf_idf_values[i])
    plt.title(f"TF-IDF values for document {i+1}: {doc}")
    plt.xlabel("Features")
    plt.ylabel("TF-IDF")
    plt.show()
````
